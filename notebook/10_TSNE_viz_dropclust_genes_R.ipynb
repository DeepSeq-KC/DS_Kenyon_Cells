```r
library(ggplot2)
library(dplyr)
```

```r
df <- read.csv("../data/07_dropclust_data/dc_gene_TSNE.csv")
```

```r
df %>%
  ggplot() +
  geom_point(aes(x=x, y=y, color=subtype))
```

```r
df %>%
  ggplot() +
  geom_point(aes(x=x, y=y, color=factor(hdb_clust)))
```

```r
cluster_tsne <- df %>%
  ggplot() +
  geom_point(aes(x=x, y=y, color=factor(hdb_clust))) +
  theme_classic() +
  xlab("TSNE Axis-1") +
  ylab("TSNE Axis-2") +
  labs(color="Cluster")

ggsave(
  filename = "../data/figures/KC_dropclust_TSNE_by_cluster.pdf",
  plot = cluster_tsne,
  width = 11,
  height = 8.5,
  units='in'
)
```

```r
df %>%
  ggplot() +
  geom_point(aes(x=x, y=y, color=factor(icim_clust)))
```

```r
icim_cluster_tsne <- df %>%
  ggplot() +
  geom_point(aes(x=x, y=y, color=factor(icim_clust))) +
  theme_classic() +
  xlab("TSNE Axis-1") +
  ylab("TSNE Axis-2") +
  labs(color="Cluster")

ggsave(
  filename = "../data/figures/KC_dropclust_TSNE_by_icim_cluster.pdf",
  plot = icim_cluster_tsne,
  width = 11,
  height = 8.5,
  units='in'
)
```
