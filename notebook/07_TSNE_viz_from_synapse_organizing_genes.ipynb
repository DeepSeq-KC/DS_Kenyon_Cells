```r
library(ggplot2)
library(dplyr)

df <- read.csv("../data/05_synapse_organizing_gene_TSNE_gen/KC_synapse_organizing_gene_TSNE_data.csv")
```

```r
df %>%
  ggplot() +
  geom_point(aes(x=x, y=y, color=factor(cell_type_id)))
```

```r
lobe_tsne <- df %>%
  ggplot() +
  geom_point(aes(x=x, y=y, color=subtype)) +
  theme_classic() +
  xlab("TSNE Axis-1") +
  ylab("TSNE Axis-2") +
  labs(color="KC Lobe")

ggsave(
  filename = "../data/figures/KC_synaptic_gene_TSNE_by_KC_lobe.pdf",
  plot = lobe_tsne,
  width = 11,
  height = 8.5,
  units='in'
)
```
